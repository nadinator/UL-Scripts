```python
# Create the directories for the splits
from pathlib import Path as P

splits = [
    "Split 2",
    "Split 3",
    "Split 4",
    "Split 5",
]

P("./Splits").mkdir(exist_ok=True)

for split in splits:
    sp = P(f"./Splits/{split}")
    sp.mkdir(exist_ok=True)
```

```python
import json

with open("./Clustrering Json/videos2images.json") as f:
    v2i = json.load(f)
    
v2i[0]
```

```python
def get_images(video):
    obj = list(filter(lambda x: x["video_name"] == video, v2i))
    assert len(obj) == 1
    return obj[0]["images"]

get_images(v2i[0]["video_name"])
```

```python
# Copy images to their split folder
from shutil import copyfile
from tqdm import tqdm
from pathlib import Path as P

split_folders = list(P("./Splits").iterdir())
for sf in tqdm(split_folders, desc="Iterating split folders..."):
    (sf / "train/").mkdir(exist_ok=True)
    (sf / "val/").mkdir(exist_ok=True)
    (sf / "test/").mkdir(exist_ok=True)
    
    for s in ["train", "val", "test"]:
        with open(sf/f"{s}.txt", encoding="utf-8") as f:
            videos = f.read().split("\n")
        for v in tqdm(videos, desc=f"Iterating videos for split {s}"):
            for image in tqdm(get_images(v), desc=f"Copying images for video {v}"):
                copyfile(P("./images")/image, sf/s/image)
                #! Forgot to copy labels!!!!
        
```
