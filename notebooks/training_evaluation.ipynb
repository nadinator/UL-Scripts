```python
from pathlib import Path as P
from pprint import pprint as pp

cwd = P()
results_csvs = list(cwd.glob("training/*/*.csv"))
print(len(results_csvs))
results_csvs
```

```python
import pandas as pd

last_rows = []

for csv in results_csvs:
    df = pd.read_csv(csv)
    last_epoch = df.iloc[-1].copy()
    last_rows += [last_epoch]

df_last_rows = pd.DataFrame(last_rows).reset_index()
df_last_rows.insert(0, 'run', [csv.parent.name for csv in results_csvs])
df_last_rows.info()
df_last_rows
```

```python
print(df_last_rows.sort_values(by="       metrics/mAP50(B)", ascending=False).head().to_markdown())
```

```python
print(df_last_rows.sort_values(by="      metrics/recall(B)", ascending=False).head().to_markdown())
```

```python
df_last_rows.sort_values(by="   metrics/precision(B)", ascending=False).head().to_csv("precision_sorted.csv")
```
